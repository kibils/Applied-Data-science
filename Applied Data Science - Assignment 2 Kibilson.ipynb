```python
import pandas as pd
import wbgapi as wb
```

```python
GDP_indcators = ['SL.UEM.1524.FE.ZS','NE.DAB.TOTL.ZS','SL.UEM.1524.NE.ZS','NE.IMP.GNFS.ZS']
cnt_cde = ["PAK","BRA",'DNK','AUS','LUX','ESP','BMU','CHE','JAM','GBR']
CLI_indcators=['EN.ATM.GHGT.KT.CE','EN.ATM.CO2E.GF.KT','EG.ELC.NGAS.ZS','EG.ELC.ACCS.RU.ZS']

#SL.UEM.1524.FE.ZS: Unemployment, female total as a %
#NE.DAB.TOTL.ZS: Expenditure total
#SL.UEM.1524.NE.ZS: Unemployment, youth total as a %
#NE.IMP.GNFS.ZS: Import total
#EN.ATM.GHGT.KT.CE: Greenshouse gas emission rate
#EN.ATM.CO2E.GF.KT: Gaseous fuel consumption CO2 emissions 
#EG.ELC.NGAS.ZS: Electricity production from natural gas sources as a % of total
#EG.ELC.ACCS.RU.ZS: Rural access to electricity
```

```python
def preprocess(g_ind,c_ind,code):
    
    '''
    Extract dataset from Worldbank website and restructure it finally return analysis ready DataFrame as output
    '''
    
    GDP = wb.data.DataFrame(g_ind, code, mrv=7)
    GDP.columns = [a.replace('YR','') for a in GDP.columns]      
    GDP = GDP.stack().unstack(level=1)                             
    GDP.index.names = ['Cnt_Cde', 'Year']
    
    CLI = wb.data.DataFrame(c_ind, code, mrv=7)
    CLI.columns = [b.replace('YR','') for b in CLI.columns]      
    CLI=CLI.stack().unstack(level=1)                             
    CLI.index.names = ['Cnt_Cde', 'Year']
    
    d1=GDP.reset_index()
    d2=CLI.reset_index()
    d3=d1.fillna(0)
    d4=d2.fillna(0)
    
    d_fnl = pd.merge(d3, d4)
    
    return d_fnl
```

```python
d_fnl = preprocess(GDP_indcators,CLI_indcators,cnt_cde)
d_fnl.head(20)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def wb_pie(var):
    '''
    This function calculates percentage of specific(input) variable with 
    Numpy tool and cxreate a new dataset that will be used to generate pie chart dynamically.
    '''
    
    df1 = d_fnl[['Cnt_Cde','Year',var]]
    df2 = df1.groupby('Year').sum().reset_index()

    total = np.sum(df2[var])

    df2['percent_np'] = np.divide(df2[var],total) * 100
    
    plt.pie(df2[var], labels=df2['Year'],autopct='%1.1f%%',startangle=90)
    plt.title('Yearwise Proportion of Unemployed Female population ')
    plt.show()
```

```python
wb_pie('SL.UEM.1524.FE.ZS')
```

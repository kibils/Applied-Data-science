```python
import numpy as np
import pandas as pd
from plotnine import geom_line, geom_point, labs, ggplot, aes, geom_boxplot
```

```python
#Reading the file
a=pd.read_csv(r"E:\\Kibilson\\Lerwick.csv", low_memory=False)
```

```python
#Scatter plot for maximum temperature
(
    ggplot(a)
    + aes(x="Years", y="T_Max", color="M")
    + labs(
        title="Maximum temperature in Lerwick",
        x="Year of analysis",
        y="Maximum Temperature in Celsius",
        color="M",fill="blue"
        
    )
    + geom_point()
)
```

```python
#Scatter plot for minimum temperature
(
    ggplot(a)
    + aes(x="Years", y="T_Min", color="M")
    + labs(
        title="Minimum temperature in Lerwick",
        x="Year of analysis",
        y="Minimum Temperature in Celsius",
        color="M",fill="blue"
        
    )
    + geom_point()
)
```

```python
#Understand rainfall pattern in Lerwick
ggplot(a, aes(x="M", y="Rain")) +geom_boxplot(fill='red')
```
